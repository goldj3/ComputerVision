## Introduction

This project was based on the article by Oliva, Torralba, and Schyns, who proposed that if the high frequency components of an image is superimposed with the low frequency components of another image, a hybrid image is formed. A hybrid image is defined by the existence of "two coherent global image interpretations" in one image. Essentially, the image with its higher frequency components present in the hybrid image will be more dominantly visible for a viewer observing from a short distance. When a viewer increases his/her distance from the picture, the image with the higher frequencies kept in the hybrid image will be more apparent. Thus, the goal of this project was to create hybrid images of our own. In order to do so, we created a high pass filter to keep the high frequencies of one image, and a low pass filter to keep the other frequency of another. The creation of these filters was possible through the use of convolution. The specific implementation of the filters and convolution will be discussed in an upcoming cell of this notebook. In the following section, there are two images which will be used to create the first hybrid image.

```python
import cv2
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from utils import vis_hybrid_image, load_image, save_image
from project2 import my_imfilter, create_hybrid_image
%load_ext autoreload
%autoreload 2

image1 = load_image('../project2/images/dog.bmp')
image2 = load_image('../project2/images/cat.bmp')

# display the dog and cat images
plt.figure(figsize=(3,3)); plt.imshow((image1*255).astype(np.uint8));
plt.figure(figsize=(3,3)); plt.imshow((image2*255).astype(np.uint8));
```

## Create filter

In the following cell, we have a gaussian kernel acting as the filter.

```python
cutoff_frequency = 7
filter = cv2.getGaussianKernel(ksize=cutoff_frequency*4+1,
                               sigma=cutoff_frequency)
filter = np.dot(filter, filter.T)

# let's take a look at the filter!
plt.figure(figsize=(4,4)); plt.imshow(filter);
```

## Algorithm Discussion

The bulk of this project was dependent on two main algorithms: my_imfilter() and create_hybrid_image().
My_imfilter() accepts either a grayscale or a colored .bmp image and a filter of two odd numbered dimensions, and returns a new image with the filter applied to the original image. This algorithm first checks the number of dimensions of the image. If the image is colored, the algorithm will separate each pixel of the image into red, green, and blue channels using the functions constructed in Project 1. Next, each channel is individually padded with values surrounding the edges of the matrix. Each channel is then convolved with the filter using my own implementation of a convolve function. Lastly, all of the channels are brought back together in the shape of the original image (using numpy.dstack and numpy.reshape for simplicity) to form the new filtered image. If the image is greyscale, the process will be exactly the same except each pixel is not broken into three channels and brought together at the end. To elaborate more on the convolve function, this function accepts a padded channel, the original channel, and a filter. The function iterates through the number of rows and the number of columns of the original channel. (Note the iteration is not through the padded channel, because the only pixel values we are concerned with adding a filter are in the original image, and not values in the padded one). As the function iterates through, a matrix with the same size of the filter is taken from the padded image. This was utilized to avoid having to iterate through each value of each channel, which would have had to be done more than once to account for overlaps of values for each convolution matrix. Then, by using the numpy.multiply function on the filter and the matrix in the padded image, all of the appropriate values of each matrix are multiplied with numpy.multiply. The sum is then taken of all the multiplied values in both arrays to get the new filtered value for one of the values in one of the pixel channels. 

The function create_hybrid_image() accepts two images and a filter, and returns the low frequency components of the first image, the high frequency components of the second image, and a hybrid image of both of the images superimposed upon each other. This is achieved by running the previous function my_imfilter individually on each image, with the same filter as the second parameter for both images. The end result is the low frequency for each image, which is what is desired from the first image. To acquire the high frequency of the second image, the new low frequency image generated from my_imfilter() is subtracted from the original image using np.subtract. The function np.clip is used on both the low and high frequency images to make sure the values are between 0 and 1.

In the following section, we use the im_filter function to acquire the low images of the dog, which creates a blurry effect.

## Create hybrid image
Next, implement _create_\__hybrid_\__image_, which takes two images and makes a hybrid image using the low frequency content from one image and the high frequency content from another.

```python
low_frequencies, high_frequencies, hybrid_image = create_hybrid_image(image1, image2, filter)
vis = vis_hybrid_image(hybrid_image)
```
