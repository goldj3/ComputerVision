# Difference of Gaussians

A common computer vision process is finding the Difference of Gaussians of an image. This is used in SIFT and some forms of blob detection. The Difference of Gaussians is found by first creating a Gaussian pyramid of an image.  Multiple 'octaves' of the pyramid are created by creating a few downsampled versions of the original image. In each octave, five different sigma values are considered by increasing the exponent of $k$ multiplied with $\sigma$ where $\sigma$ = 1.6 and $k$ = $\sqrt2$ . In other words, the $\sigma$ for each Gaussian filter is generated by: $\sigma$, $\sigma$$k$, $\sigma$$k^2$, $\sigma$$k^3$, $\sigma$$k^4$. After creating these increasingly downsampled and blurred versions of the image, new levels for each octave are created by subtracting the higher filtered components from the lesser filtered components.  

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage.transform import rescale, resize, downscale_local_mean
import math
```

### Generating Four Octaves of Different Gaussian Filters and Scales to make Gaussian Pyramid

```python
img = cv2.imread('../portfolio/images/nd1.jpg')

sigma = 1.6
k = math.sqrt(2)

gray = np.array(cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)) # full image
gray = np.float64(gray)

gray1 = cv2.GaussianBlur(gray,(5,5),sigma,sigma)
gray2 = cv2.GaussianBlur(gray,(7,7),sigma*k,sigma*k)
gray3 = cv2.GaussianBlur(gray,(9,9),sigma*(k**2),sigma*(k**2))
gray4 = cv2.GaussianBlur(gray,(11,11),sigma*(k**3),sigma*(k**3))


half = rescale(gray, 0.5, anti_aliasing=False,multichannel=False, mode='constant') # Image downsampled by 2

half1 = cv2.GaussianBlur(half,(5,5),sigma,sigma)
half2 = cv2.GaussianBlur(half,(7,7),sigma*k,sigma*k)
half3 = cv2.GaussianBlur(half,(9,9),sigma*(k**2),sigma*(k**2))
half4 = cv2.GaussianBlur(half,(11,11),sigma*(k**3),sigma*(k**3))


quarter = rescale(gray, 0.25, anti_aliasing=False,multichannel=False, mode='constant') # Image downsampled by 4

quarter1 = cv2.GaussianBlur(quarter,(5,5),sigma,sigma)
quarter2 = cv2.GaussianBlur(quarter,(7,7),sigma*k,sigma*k)
quarter3 = cv2.GaussianBlur(quarter,(9,9),sigma*(k**2),sigma*(k**2))
quarter4 = cv2.GaussianBlur(quarter,(11,11),sigma*(k**3),sigma*(k**3))



eighth = rescale(gray, 0.125, anti_aliasing=False,multichannel=False, mode='constant') # Image downsampled by 8

eighth1 = cv2.GaussianBlur(eighth,(5,5),sigma,sigma)
eighth2 = cv2.GaussianBlur(eighth,(7,7),sigma*k,sigma*k)
eighth3 = cv2.GaussianBlur(eighth,(9,9),sigma*(k**2),sigma*(k**2))
eighth4 = cv2.GaussianBlur(eighth,(11,11),sigma*(k**3),sigma*(k**3))


print('Resolution of Full Image:', gray.shape)
print('Resolution of image downsampled by factor of 2:', half.shape)
print('Resolution of image downsampled by factor of 4:',quarter.shape)
print('Resolution of image downsampled by factor of 8:', eighth.shape)
```

### Calculating Difference of Gaussians for Each Octave

```python
# Octave 1
oct1_1 = gray2 - gray1
oct1_2 = gray3 - gray2
oct1_3 = gray4 - gray3

# Octave 2
oct2_1 = half2 - half1
oct2_2 = half3 - half2
oct2_3 = half4 - half3

# Octave 3
oct3_1 = quarter2 - quarter1
oct3_2 = quarter3 - quarter2
oct3_3 = quarter4 - quarter3

# Octave 4
oct4_1 = eighth2 - eighth1
oct4_2 = eighth3 - eighth2
oct4_3 = eighth4 - eighth3
```
