# SIFT Detection

Not only does David Lowe describe in his research a method for finding keypoint descriptors, he additionally discusses an alternative way of feature detection. After looking through various resources for further explanation of the SIFT Descriptor, in every one what is previously mentioned is a SIFT Detector. While, the Harris Detector is rotationally invariant, it is not scale invariant.   I attempt to create an implementation of the SIFT Detector based off the following resources:


* https://www.cs.ubc.ca/~lowe/papers/ijcv04.pdf

* https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_feature2d/py_sift_intro/py_sift_intro.html


As David Lowe describes in his paper, there are four steps in SIFT or Scale Invariant Feature Transform:

1. Scale-Space Extrema Detection
2. Keypoint Localization
3. Orientation Assignment
4. Keypoint Descriptor

As step 4 is covered in Project 5 I attept to recreate steps 1 and 2.

## 1. Scale-Space Extrema Detection

The first step regarding the SIFT Detector is taking the considering different Gaussian filters on an image with varying kernel sizes and sigma values. Multiple 'octaves' are created for each downsampled version of the image. In each octave, five different sigma values are considered by increasing the exponent of $k$ multiplied with $\sigma$ where $\sigma$ = 1.6 and $k$ = $\sqrt2$ . In other words, the $\sigma$ for each Gaussian filter is generated by: $\sigma$, $\sigma$$k$, $\sigma$$k^2$, $\sigma$$k^3$, $\sigma$$k^4$. Then, the different images are 

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage.transform import rescale, resize, downscale_local_mean
import math
```

### Generating Four Octaves of Different Gaussian Filters and Scales

```python
img = cv2.imread('../portfolio/images/nd1.jpg')

sigma = 1.6
k = math.sqrt(2)

gray = np.array(cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)) # full image
gray = np.float64(gray)

gray1 = cv2.GaussianBlur(gray,(5,5),sigma,sigma)
gray2 = cv2.GaussianBlur(gray,(7,7),sigma*k,sigma*k)
gray3 = cv2.GaussianBlur(gray,(9,9),sigma*(k**2),sigma*(k**2))
gray4 = cv2.GaussianBlur(gray,(11,11),sigma*(k**3),sigma*(k**3))


half = rescale(gray, 0.5, anti_aliasing=False,multichannel=False, mode='constant') # Image downsampled by 2

half1 = cv2.GaussianBlur(half,(5,5),sigma,sigma)
half2 = cv2.GaussianBlur(half,(7,7),sigma*k,sigma*k)
half3 = cv2.GaussianBlur(half,(9,9),sigma*(k**2),sigma*(k**2))
half4 = cv2.GaussianBlur(half,(11,11),sigma*(k**3),sigma*(k**3))


quarter = rescale(gray, 0.25, anti_aliasing=False,multichannel=False, mode='constant') # Image downsampled by 4

quarter1 = cv2.GaussianBlur(quarter,(5,5),sigma,sigma)
quarter2 = cv2.GaussianBlur(quarter,(7,7),sigma*k,sigma*k)
quarter3 = cv2.GaussianBlur(quarter,(9,9),sigma*(k**2),sigma*(k**2))
quarter4 = cv2.GaussianBlur(quarter,(11,11),sigma*(k**3),sigma*(k**3))



eighth = rescale(gray, 0.125, anti_aliasing=False,multichannel=False, mode='constant') # Image downsampled by 8

eighth1 = cv2.GaussianBlur(eighth,(5,5),sigma,sigma)
eighth2 = cv2.GaussianBlur(eighth,(7,7),sigma*k,sigma*k)
eighth3 = cv2.GaussianBlur(eighth,(9,9),sigma*(k**2),sigma*(k**2))
eighth4 = cv2.GaussianBlur(eighth,(11,11),sigma*(k**3),sigma*(k**3))


print('Resolution of Full Image:', gray.shape)
print('Resolution of image downsampled by factor of 2:', half.shape)
print('Resolution of image downsampled by factor of 4:',quarter.shape)
print('Resolution of image downsampled by factor of 8:', eighth.shape)
```

With octaves containing the difference of Gaussians, local extrema can be found by comparing each pixel to its 8 neighbors on its own scale, the 9 neighbors (including its corresponding pixel position) on the scales above and below the current scale. 

```python
def localExtrema(scale,scaleAbove,scaleBelow):
    scaleAbove = np.pad(oct1_1, ((1,1),(1,1)), 'constant',constant_values=0)
    currScale = np.pad(oct1_2, ((1,1),(1,1)), 'constant',constant_values=0)
    scaleBelow = np.pad(oct1_3, ((1,1),(1,1)), 'constant',constant_values=0)
    scaleAct = np.copy(oct1_2)
    extrema = []
    for i in range(scaleAct.shape[0]):
        for j in range(scaleAct.shape[1]):
            allNeighbors = np.array([
            currScale[i+1,j+1],
            currScale[i+1,j],
            currScale[i+1,j+2],
            currScale[i,j+2],
            currScale[i+2,j],
            currScale[i,j+1],
            currScale[i+2,j+1],
            currScale[i,j],
            currScale[i+2,j+2],
            scaleAbove[i+1,j],
            scaleAbove[i+1,j+2],
            scaleAbove[i,j+2],
            scaleAbove[i+2,j],
            scaleAbove[i,j+1],
            scaleAbove[i+2,j+1],
            scaleAbove[i,j],
            scaleAbove[i+2,j+2],
            scaleAbove[i+1,j+1],
            scaleBelow[i+1,j],
            scaleBelow[i+1,j+2],
            scaleBelow[i,j+2],
            scaleBelow[i+2,j],
            scaleBelow[i,j+1],
            scaleBelow[i+2,j+1],
            scaleBelow[i,j],
            scaleBelow[i+2,j+2],
            scaleBelow[i+1,j+1],
            ])    
            if(np.max(allNeighbors) == currScale[i+1,j+1] or np.min(allNeighbors) == currScale[i+1,j+1]):
                extrema.append(currScale[i+1,j+1])
    return extrema
            
        
        

```
