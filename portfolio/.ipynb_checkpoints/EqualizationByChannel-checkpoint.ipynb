## Equalizing Each Color Channel of Zebra Image

```python

#Load original image
rgb = loadppm("../images/zebra.ascii.ppm")
plt.xticks([]), plt.yticks([])   # to hide tick values on X and Y axis
plt.title('Original Zebra Image')
plt.imshow(rgb)
plt.show()

# Generate equalized color channels
red = GetRedPixels(rgb)
green = GetGreenPixels(rgb)
blue = GetBluePixels(rgb)

red_equalized = rescale(red)
green_equalized = rescale(green)
blue_equalized = rescale(blue)
```

```python
## Showing new equalized color channels

plt.xticks([]), plt.yticks([])   # to hide tick values on X and Y axis
plt.title('Red Channel Equalized')
plt.imshow(red_equalized)
plt.show()
plt.title('Green Channel Equalized')
plt.xticks([]), plt.yticks([])   # to hide tick values on X and Y axis
plt.imshow(green_equalized)
plt.show()
plt.title('Blue Channel Equalized')
plt.xticks([]), plt.yticks([])   # to hide tick values on X and Y axis
plt.imshow(blue_equalized)
plt.show()
```
