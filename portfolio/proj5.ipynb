## Our Writeup:


###  Interest Point Detection:
Interest points are detected using an optimized implementation of the Harris feature detector. We employ non-maximum suppression over detected corners greater than 1\% of the maximum r. Optimizations include vectorization of distance calculations of NMS. We ultimately select the 2000 best detected points by default.

### Feature Descriptors:
Our feature descriptor uses most of the components described in Lowe's SIFT paper. We applied a Gaussian filter to blur to the image, and then used the Sobel Filters to calculate the x and y gradients. Then, by using np.arctan() we found the orientation of the image. We created a 16x16 window around each feature point, and divided the window into 4x4 subregions. We made an orientation histogram for each of the subregions, and used the histograms of each subregion as our feature vector. To improve on the descriptor, we implemented a few additional processes that were described in the SIFT paper. We normalized the final feature vector to unit length, as Lowe indicates doing so will reduce affine illumination changes. By thresholding each value to be less than 0.2 and renormalizing the feature vector to unit length, the larger gradients would have not been as significant in feature matching, and instead the more significant factor would be the distribution of the orientation values among the features. However, upon implementing these techniques, the accuracy was either decreased or was the same as our basic descriptor. Another technique we implemented that was suggested raising each value in our descriptor to a power <1. In doing so, we found a range that increased the accuracy by a few percentage points. We found the range that contributed to increasing the accuracy the most was between 0.5-0.9. 

### Feature Matching:
We implemented NNDR (The nearest neighbor distance ratio test) which attempts to classify if a matching between two feature vector is good based on the euclidian distance between the feature vectors. For each feature point in image 1, it finds the best two matches in image 2, then if the matches are too similar it ignores them.
